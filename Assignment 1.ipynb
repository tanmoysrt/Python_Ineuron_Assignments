#### 2.
#### Create the below pattern using nested for loop in Python.
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

```python
for i in range(1,6,1):
    for j in range(i):
        print('*',end=' ')
    print()
for i in range(4,0,-1):
    for j in range(i):
        print('*',end=' ')
    print('')
```
